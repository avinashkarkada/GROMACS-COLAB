```python
%%bash
source /content/gromacs/bin/GMXRC
gmx editconf -f complex.gro -o newbox.gro -bt cubic -d 2.0 -c
gmx grompp -f vacuum.mdp -c newbox.gro -p topol.top -o protein-EM-vacuum.tpr -maxwarn 200
gmx mdrun -v -deffnm protein-EM-vacuum -c protein-EM-vacuum.pdb
gmx mdrun -v -deffnm protein-EM-vacuum
```

```python
%%bash
source /content/gromacs/bin/GMXRC
gmx solvate -cp protein-EM-vacuum.gro -cs spc216.gro -p topol.top -o solv.gro
gmx grompp -f em.mdp -c solv.gro -p topol.top -o ions.tpr -maxwarn 2
gmx genion -s ions.tpr -o solv_ions.gro -p topol.top -conc 0.15  -neutral -pname NA -nname CL <<< 15
gmx grompp -f em.mdp -c solv_ions.gro -p topol.top -o em.tpr -maxwarn 2
gmx mdrun -v -deffnm em
```

```python
%%bash
source /content/gromacs/bin/GMXRC
gmx genrestr -f ligand.gro -o posre_ligand.itp -fc 1000 1000 1000 <<< 0
sed -i '/Include water topology/i \; Include Position restraint file' topol.top
sed -i '/Include water topology/i \#ifdef POSRES' topol.top
sed -i '/Include water topology/i \#include "posre_ligand.itp"' topol.top
sed -i '/Include water topology/i \#endif' topol.top
sed -i '/Include water topology/i \ ' topol.top
printf "%s\n" "1 | 13" "q" |  gmx make_ndx -f em.gro -o index.ndx 
```

```python
%%bash
source /content/gromacs/bin/GMXRC

arr=(*_*_*_*.itp) && echo "${arr[0]}"

filename=${arr[0]}

if [ -f "${filename}" ]

then

vim -c ":%s/1000  1000  1000/100     100     100/g" -c wq <<< E ${arr[0]}
vim -c ":%s/1000  1000  1000/100     100     100/g" -c wq <<< E ${arr[1]}
vim -c ":%s/1000       1000       1000/100     100     100/g" -c wq <<< E posre_ligand.itp

gmx grompp -f npt.mdp -c npt-1000.gro -t npt-1000.cpt -p topol.top -n index.ndx -o npt-100.tpr -r em.gro -maxwarn 2 
gmx mdrun -deffnm npt-100  -v
vim -c ":%s/100     100     100/10     10     10/g" -c wq <<< E ${arr[0]}
vim -c ":%s/100     100     100/10     10     10/g" -c wq <<< E ${arr[1]}
vim -c ":%s/100     100     100/10     10     10/g" -c wq <<< E posre_ligand.itp

gmx grompp -f npt.mdp -c npt-100.gro -t npt-100.cpt -p topol.top -n index.ndx -o npt-10.tpr -r em.gro -maxwarn 2 
gmx mdrun -deffnm npt-10  -v
vim -c ":%s/10     10     10/1     1     1/g" -c wq <<< E ${arr[0]}
vim -c ":%s/10     10     10/1     1     1/g" -c wq <<< E ${arr[1]}
vim -c ":%s/10     10     10/1     1     1/g" -c wq <<< E posre_ligand.itp


else 
vim -c ":%s/1000  1000  1000/100     100     100/g" -c wq <<< E posre.itp
vim -c ":%s/1000       1000       1000/100     100     100/g" -c wq <<< E posre_ligand.itp

gmx grompp -f npt.mdp -c npt-1000.gro -t npt-1000.cpt -p topol.top -n index.ndx -o npt-100.tpr -r em.gro -maxwarn 2 
gmx mdrun -deffnm npt-100  -v
vim -c ":%s/100     100     100/10     10     10/g" -c wq <<< E posre.itp
vim -c ":%s/100     100     100/10     10     10/g" -c wq <<< E posre_ligand.itp

gmx grompp -f npt.mdp -c npt-100.gro -t npt-100.cpt -p topol.top -n index.ndx -o npt-10.tpr -r em.gro -maxwarn 2 
gmx mdrun -deffnm npt-10  -v
vim -c ":%s/10     10     10/1     1     1/g" -c wq <<< E posre.itp
vim -c ":%s/10     10     10/1     1     1/g" -c wq <<< E posre_ligand.itp
fi


gmx grompp -f npt.mdp -c npt-10.gro -t npt-10.cpt -p topol.top -n index.ndx -o npt-1.tpr -r em.gro -maxwarn 2 
gmx mdrun -deffnm npt-1 -v

vim -c ":%s/define/;define/g" -c wq <<< E npt.mdp

gmx grompp -f npt.mdp -c npt-1.gro -t npt-1.cpt -p topol.top -n index.ndx -o npt.tpr -r em.gro -maxwarn 2
gmx mdrun -deffnm npt -v
```

```python
#@title Analysis
%%bash
source /content/gromacs/bin/GMXRC
printf '1\n1' | gmx trjconv -s protein-ligand-out.tpr -f protein-ligand-out.xtc -o protein-ligand-out.pdb -pbc cluster
printf '1\n1' | gmx trjconv -s protein-ligand-out.tpr -f protein-ligand-out.xtc -o protein-0-10ns.xtc -pbc cluster
printf '4\n4' | gmx rms -s protein-ligand-out.tpr -f protein-ligand-out.pdb -o protein-0-1-RMSD.xvg 
gmx rmsf -s protein-ligand-out.tpr -f protein-ligand-out.pdb -o protein-0-10ns-RMSF.xvg -res <<< 3
```
